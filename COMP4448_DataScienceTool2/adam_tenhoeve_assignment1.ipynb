Imagine that you wanted to use the student_data in question 2a to make predictions such that the ice_cream_flavor, math and reading columns are input variables and science column is the output variable you want to predict. 

Part a) Use the LabelBinarizer() in the sklearn package to transform the ice_cream_flavor column in the student_data to dummy variables, then join these dummy variables to the student_data and drop the original ice_cream_flavor column. Reassign the resulting DataFrame to a variable called student_data_1. Print out the entire student_data_1 DataFrame 

```python
from sklearn.preprocessing import LabelBinarizer
# Turn the ice cream flavor into dummy variables
lb = LabelBinarizer()
flavors_dummy = lb.fit_transform(student_data["ice_cream_flavor"])
flavors_dummy = pd.DataFrame(flavors_dummy, columns=lb.classes_)
# Join the dummy variables to the student data
student_data_1 = student_data.join(flavors_dummy)
# Drop the original ice_cream_flavor column
student_data_1.drop("ice_cream_flavor", axis=1, inplace=True)
student_data_1
```

Part B) Extract the math, reading and science scores and use the StandardScaler() class in the sklearn.preprocessing module to standardize these scores, then merge the standardized scores to the dummy variables, and call the resulting DataFrame student_data_std. Print out the entire student_data_1 DataFrame. 

```python
from sklearn.preprocessing import StandardScaler
# Use the standard scaler to standarize the numeric values
scaler = StandardScaler()
scaled_scores = scaler.fit_transform(student_data_1[["math", "reading", "science "]])
scaled_scores = pd.DataFrame(scaled_scores, columns=["math", "reading", "science"])
# Join the scaled scores and the dummy variables
student_data_std = scaled_scores.join(flavors_dummy)
student_data_std
```

Part D) Using the boston_1 DataFrame in question 1a, select the LowTemp, HighTemp, WarmestMin, ColdestHigh, AveMin, AveMax columns and use the “pipeline” functionality in sklearn to transform the selected data using the MinMaxScalar() and Imputer() classes.  With the Imputer() class, the missing data in each column should be imputed using the mean value for the column. Assign the resulting DataFrame to a variable called pipeline_data. Print the pipeline_data

```python
# Select desired columns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer

# Create the pipeline
pipe = Pipeline([("scaler", MinMaxScaler()), 
                 ("impute", SimpleImputer(strategy="mean"))])
# Fit the pipeline to the specified columns of the boston_1 data
specified_cols = ["LowTemp", "HighTemp", "WarmestMin", "ColdestHigh", "AveMin", "AveMax"]
pipeline_data = pipe.fit_transform(boston_1[specified_cols])
pipeline_data = pd.DataFrame(pipeline_data, columns=specified_cols)
pipeline_data
```
